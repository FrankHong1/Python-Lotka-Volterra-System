```python
def euler_lotka_volterra(h,t,tspan,x_initial,y_initial,params):
    # x,y,t             --  the initial conditions
    # tspan             -- time span
    # h                 -- size of the step
    # params            -- array of parameters,there is supposed to have 4 parameters
    #       a   - birth rate of the rabbit
    #       b   - death rate of rabbits due to foxes
    #       c   - is the birth rate of foxes
    #       d   - is the death rate of foxes.
    X=[x_initial]
    Y=[y_initial]
    N = tspan/h+1
    T=np.linspace(t,tspan,N)

    a,b,c,d=params[0],params[1],params[2],params[3]

    x_previous = x_initial
    y_previous = y_initial
    for k in range(0,int(tspan/h)):
        # calculate values using Euler forward method in a for loop
        x_next=(x_previous)+h*(a*x_previous-(b*x_previous*y_previous))
        y_next=(y_previous)+h*(c*x_previous*y_previous-(d*y_previous))

        x_previous=x_next
        y_previous=y_next
        X.append(x_previous)
        Y.append(y_previous)
    plt.plot(T,X,'r',label='rabbits')
    plt.plot(T,Y,'b',label='foxes')
    plt.xlabel('time')
    plt.title('Predator prey model')
    plt.legend()
    plt.show #Plotting the result on the graph
    print('the number of rabbits at t=',t,'are',x,', the number of foxes are',y)
    return None
```

```python
h=0.001
t=0
tspan=1
x=.9
y=.1
params=[1,0.4,0.5,0.1]
euler_lotka_volterra(h,t,tspan,x,y,params)
#In this case the foxes has too much advantages in the parameters, and thus both of them distinguished at the end.
```
