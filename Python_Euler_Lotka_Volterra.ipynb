```python
import matplotlib.pyplot as plt
```

```python
def euler_lotka_volterra(h,t,tspan,x,y,params):
    #x,y,t are the initial conditions,tspan is the time span, h is the size of the step, params is an array of parameters,there is supposed 
    #to have 4 parameters, a is the birth rate of the rabbit, b is the eating rate of foxes, c is the birth rate of foxes, d is the death 
    #rate of foxes.
    X=[]
    X.append(x)
    Y=[]
    Y.append(y)
    T=[]
    T.append(t)
    a=params[0]
    b=params[1]
    c=params[2]
    d=params[3]
    for k in range(0,int(tspan/h)):
        #predict values using euler forward method in a for loop
        x_next=(x)+h*(a*x-(b*x*y))
        y_next=(y)+h*(c*x*y-(d*y))
        t=t+h
        T.append(t)
        x=x_next
        y=y_next
        X.append(x)
        Y.append(y)
    plt.plot(T,X,'r')
    plt.plot(T,Y,'b')
    plt.xlabel('time')
    plt.title('Predator prey model')
    plt.show #Plotting the result on the graph
    print('the number of rabbits at t=',t,'are',x,', the number of foxes are',y)
    return None
```

```python
h=0.001
t=0
tspan=200
x=200
y=10
params=[1,0.4,0.5,0.1]
euler_lotka_volterra(h,t,tspan,x,y,params)
#In this case the foxes has too much advantages in the parameters, and thus both of them distinguished at the end.
```
